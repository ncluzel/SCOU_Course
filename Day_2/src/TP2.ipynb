```python
def cross_correlation_matcher_E2(s1, s2, max_lag):
    lags = np.arange(-max_lag, max_lag+1)
    corrs = []
    for lag in lags:

        if lag==0:
            sub_s1 = s1
            sub_s2 = s2
        elif lag > 0:
            sub_s1 = s1[lag:]
            #sub_s2 = s2[...] # to complete
        else:
            #sub_s1 = #s1[...] # to complete
            #sub_s2 = #s2[...] # to complete

        corr = np.corrcoef(sub_s1, sub_s2)[0, 1]
        corrs.append(corr)

    return dict(zip(lags, corrs)), np.max(corrs), lags[np.argmax(corrs)]
```

### Exercice 3 : Evaluer l'incertitude sur l'estimation du lag

Pour obtenir une incertitude sur l'estimation du lag, on peut estimer des intervalles de confiance par sous-échantillonnage Monte Carlo.

- Q1 : Modifiez votre fonction précédente pour rajouter un sous-échantillonnage Monte-Carlo permettant d'obtenir des intervalles de confiance sur l'estimation du lag et de la corrélation. Complétez pour cela la fonction cross_correlation_matcher.

```python
def cross_correlation_matcher(s1, s2, max_lag=15, prop=0.5, B=1000, use_subsampling=True, seed=48, use_Pearson=True):

    lags = np.arange(-max_lag, max_lag+1)
    boot_lags = []
    boot_corrs = []

    n = s1.shape[0]
    complete_indexes = np.arange(0, n)
    size = int(prop * n)

    if use_subsampling:
        np.random.seed(seed)
        for _ in range(B):
            idx = np.array(sorted(np.random.choice(complete_indexes, size=size, replace=False)))
            valids = []
            # Ensuring that only valid points are taken into account and that all correlations are computed on the same signal :
            for lag in [lags[0], lags[-1]]:
                shifted = idx + lag
                valid = (shifted >= 0) & (shifted < n)
                valids.append(valid)

            valid_concat = []
            for v_idx, v1 in enumerate(valids[0]):
                v2 = valids[-1][v_idx]
                if v1 and v2:
                    valid_concat.append(True)
                else:
                    valid_concat.append(False)
            valid_concat = np.array(valid_concat)

            corrs = []
            for lag in lags:
                shifted = idx + lag
                sub_s1 = s1[idx[valid_concat]]
                #sub_s2 = s2[...] hint: if s1 stays still, then s2 has to be shifted.

                if use_Pearson:
                    #corr =  ... hint: check cross_correlation_matcher_E2 function
                else:
                    #corr =  ... find a Python function allowing to compute Spearman's r coefficient. You can find it in a popular library.
                corrs.append(corr)

            corr_max = np.max(corrs)
            #lag_max =  ... hint: if corr_max is taken from the maximum of correlation, how do we gather lag_max ?

            boot_lags.append(lag_max)
            boot_corrs.append(corr_max)

    return boot_lags, boot_corrs
```

### Exercice 5 : Evaluer l'incertitude liée à la reconstruction du signal à partir de mesures bruitées

```python
data = pd.read_csv(base_dir.split('src')[0] + '/data/PARIS.csv', sep=";")
data = data.loc[::, ['end', 'ti_metropole']]
data = data.rename(columns={'end':'dateStart', 'ti_metropole':'Xt'})
data.dateStart = pd.to_datetime(data.dateStart)
data.Xt = np.log10(data.Xt)
```

```python
data
```

```python
RANDOM_SEED = 21

# Génération d'un signal synthétique. Faisons simple : pas d'outliers ni de censure
gt_eps = 0.4

# Initialisation de la graine :
np.random.seed(RANDOM_SEED)

Yt = np.zeros(data.shape[0])
for i in range(data.shape[0]):
    Yt[i] = data.loc[i, 'Xt'] + np.random.normal(loc=0, scale=gt_eps)

df = data.copy()
df['obs'] = Yt
df['lod'] = -1000

full_data = df.copy()
```

```python
full_scou = model_fit(full_data)
```

```python
get_results(full_data, full_scou, remove_those = [])
```

```python
# Maintenant, nous allons sous-échantillonner le signal et voir comment cela affecte la reconstruction
```

```python
from itertools import combinations

elements = [0, 1, 2, 3, 4, 5, 6]

combinations = [(a, b) for a, b in combinations(elements, 2) if b - a >= 2]

print(combinations)
```

```python
subsignals = {}

for comb in combinations:
    comb = list(comb)
    sub_data = df.copy()
    sub_data['weekday'] = sub_data.dateStart.dt.weekday
    sub_data['id'] = sub_data.index.tolist()

    kept_days = [el for el in np.arange(7).tolist() if el not in comb]

    sub_data.loc[sub_data.weekday.isin(kept_days), 'id'] = np.nan
    sub_data.loc[sub_data.id.isna(), 'obs'] = np.nan

    lbound = sub_data.loc[~sub_data.obs.isna()].index.tolist()[0]
    rbound = sub_data.loc[~sub_data.obs.isna()].index.tolist()[-1]

    sub_data = sub_data.loc[lbound:rbound]
    sub_data.reset_index(inplace=True, drop=True)

    sub_scou = model_fit(sub_data)
    get_results(sub_data, sub_scou, remove_those = [])

    key = f'{comb[0]}-{comb[1]}'
    subsignals[key] = sub_data
```

```python
# Maintenant, pour chaque sous-jeu de données, on va prendre le sous-signal reconstruit, et estimer sa corrélation et surtout
# son décalage temporel avec le signal d'origine, en s'intéressant à une vague particulière (l'avant-dernière)
```

```python
def plot_signals(s1, s2):
    context = []

    with plt.style.context(context):
        LABEL_SIZE = 30
        TICK_SIZE = 30
        TITLE_SIZE = 38
        LEGEND_SIZE = 30
        DATES_SIZE = 18
        figsize = (28, 7)

        plt.rc('axes', labelsize=LABEL_SIZE)
        plt.rc('xtick', labelsize=TICK_SIZE)
        plt.rc('ytick', labelsize=TICK_SIZE)
        plt.rc('figure', titlesize=TITLE_SIZE)
        plt.rc('legend', fontsize=LEGEND_SIZE)

        fig = plt.figure(figsize=figsize, layout="constrained")
        ax_dict = fig.subplot_mosaic(
            """
            A
            """
        )

        ax_dict['A'].plot(s1.dateStart.values, s1.Xt.values, color='dodgerblue', linewidth=10, zorder=3)
        ax_dict['A'].plot(s1.dateStart.values, s1.Xt.values, color='black', linewidth=3, zorder=3)

        ax_dict['A'].plot(s2.dateStart.values, s2.muX.values, color='darkorange', linewidth=10, zorder=3)
        ax_dict['A'].plot(s2.dateStart.values, s2.muX.values, color='black', linewidth=3, zorder=3)

        ax_dict['A'].grid()

        plt.show()
```

```python
left_bound = pd.to_datetime('2021-11-01') - datetime.timedelta(days=7)
right_bound = pd.to_datetime('2022-03-03') + datetime.timedelta(days=7)

signal_1 = full_data.loc[(full_data.dateStart>=left_bound)&(full_data.dateStart<=right_bound)]
s1 = signal_1.Xt.values

for key in list(subsignals.keys()):
    sub_data = subsignals[key]
    signal_2 = sub_data.loc[(sub_data.dateStart>=left_bound)&(sub_data.dateStart<=right_bound)]
    s2 = signal_2.muX.values
    tlags, tcorrs = cross_correlation_matcher(s1, s2, prop=0.5)
    plot_single_wave(tcorrs, tlags)
    plot_signals(signal_1, signal_2)
```
