- ensuite pareil en intégrant différentes proportions de données censurées

```python
data = pd.read_csv('../data/df_Tunis.csv', sep=";")
data = data.rename(columns={'scou_muX':'muX'})
data.dateStart = pd.to_datetime(data.dateStart)
```

```python
data
```

```python
plt.plot(data.muX.values)
```

```python
X_t = data.muX.values
eps = .5
np.random.seed(48)

noise = np.random.normal(loc=0, scale=eps, size=X_t.shape[0])
Y_t = X_t + noise
```

```python
plt.figure(figsize=(16,5))
plt.plot(data.dateStart.values, data.muX.values)
plt.scatter(data.dateStart.values, Y_t)
```

```python
df = pd.DataFrame()
df['dateStart'] = data.dateStart.values
df['X_t'] = X_t
df['obs'] = Y_t
df['lod'] = -1000
```

```python
df
```

```python
for col in ['X_t', 'obs', 'lod']:
    df[col] = np.log10(np.exp(df[col]))
```

```python
for col in ['X_t', 'obs', 'lod']:
    df[col] = 10**(df[col])
```

```python
df['lod'] = 10**(-30)
```

```python
df
```

```python
df.to_csv('../data/dataset.csv', sep=";", index=False)
```
